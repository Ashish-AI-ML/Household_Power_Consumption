```python
df_one = pd.read_csv('/content/household_power_consumption.txt', sep=';',dtype=str, low_memory=False)
```

```python
df_one.head()
```

```python

# Combine 'Date' and 'Time' into a single datetime column
df_one['datetime'] = pd.to_datetime(df_one['Date'] + ' ' + df_one['Time'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
```

```python
df_one.head()
```

```python
# Drop rows where datetime parsing failed
df_one = df_one.dropna(subset=['datetime'])
```

```python
df_one.info()
```

```python
# Set 'datetime' as the index
df_one = df_one.set_index('datetime')
```

```python

# Drop original 'Date' and 'Time' columns
df_one = df_one.drop(columns=['Date', 'Time'])
```

```python
# Identify columns that should be numeric
numeric_cols = df_one.columns.tolist()  # all remaining columns should be numeric
```

```python
# Replace '?' with NaN and convert to numeric
for col in numeric_cols:
    df_one[col] = df_one[col].replace('?', np.nan)
    df_one[col] = pd.to_numeric(df_one[col], errors='coerce')
```

```python
df_one.head()
```

```python
# Rename 'Global_active_power' to 'energy' for convenience
df_one.rename(columns={'Global_active_power': 'energy'}, inplace=True)
```

```python
# Optional: Check how many NaNs exist before resampling
print(f"NaNs in 'energy': {df_one['energy'].isna().sum()}")
```

```python
# Resample by day and sum 'energy'
# This will ignore NaNs (by default), and if a full day is NaN, the sum will be 0.0
# then convert to kWh
daily_energy_kwh = df_one['energy'].resample('D').sum() / 60  # because data is per minute
```

```python
# View result
print(daily_energy_kwh.head())
```

```python
# Remove days with no data (sum == 0) or NaN sums
daily_energy_kwh = daily_energy_kwh[daily_energy_kwh > 0]
daily_energy_kwh = daily_energy_kwh.dropna()
```

```python
# Assume 'daily_energy' is a Series with datetime index and energy values
df_lagged = pd.DataFrame({'y': daily_energy_kwh})
```

```python
# Add lagged features: energy from previous 1 to 7 days
for lag in range(1, 8):
    df_lagged[f'lag_{lag}'] = df_lagged['y'].shift(lag)
```

```python
# Drop rows with any NaNs (first 7 rows will be NaN due to shifting)
df_lagged.dropna(inplace=True)
```

```python
# Optional: inspect the prepared data
print(df_lagged.head())
```

```python
# Get the number of rows for an 80/20 split
split_index = int(len(df_lagged) * 0.8)
```

```python
# Chronological split
train = df_lagged.iloc[:split_index]
test = df_lagged.iloc[split_index:]
```

```python
# Separate features and target
X_train, y_train = train.drop(columns='y'), train['y']
X_test, y_test = test.drop(columns='y'), test['y']
```

Linear Regression model

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict on test set
y_pred_lr = lr_model.predict(X_test)
```

 Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict on test set
y_pred_rf = rf_model.predict(X_test)
```

Visualize Correlation Between Lags and Target

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation of all columns including target
corr = df_lagged.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr[['y']].T, annot=True, cmap='coolwarm')
plt.title("Correlation of Lag Features with Target 'y'")
plt.show()
```

Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score


# For Linear Regression
print("Linear Regression:")
# Calculate MSE and then take the square root for RMSE
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print("R²:", r2_score(y_test, y_pred_lr))

# For Random Forest
print("\nRandom Forest:")
# Calculate MSE and then take the square root for RMSE
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("R²:", r2_score(y_test, y_pred_rf))
```
